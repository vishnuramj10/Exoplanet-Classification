### Usual Questions 
1/ What is an exoplanet - 
Exoplanet or extra solar planet is any planet which is resolving around a star 
other than our solar system. Exoplanets can vary in size, characteristics and provide the diversity of different planets in our universe. 

2/ How does Nasa get the data-
NASA collects exoplanet data using various space-based and ground-based observatories and instruments, which are designed to detect and characterize the properties of exoplanets and their host stars. Few of the observations for exoplanets are by measuring the periodic dimming of stars as planets transit, or pass in front of, their host stars. Nasa also uses the hubble space telescope to understand the exo planets even more and even ground based observatories which find exoplanets by seeing how their gravitational pull is among the host stars. They also take visual images of exoplanets.

3/ What kind of features can help NASA identify exo planets or how does one find exoplanets?
I think this is an important question for finding exoplanets. NASA uses different techniques to find an exoplanet.

a/ Transit Method: This method looks for observing the brightness of a star and its periodic dip ( indicating that a planet is transiting or passing in front) This method can give information about the size and distance of planet from host. 

b/ Radial Velocity Method: We use the gravitational pull of the planet from its orbiting star. This method can provide information about the mass and distance of the planet from its host star.

Gravitational Microlensing Method: This method involves the usage of theory of relativity that is bending of light around a massive object, such as a star or planet. Fun Fact - This is how einstein prooved theory of relativity using a solar eclipse!

Direct Imaging Method: Just taking the picture. It can provide information about the planet's size, composition, and orbit.

Astrometry Method: This method is one of the most interesting one!. The host star which the planets revolves around wobbles due to the gravitational pull of its planet. We try to detect this.

Transit Timing Variations Method: This method involves observing the variations in the timing of a planet's transits as it is gravitationally influenced by another planet in the same system.

### Handling Outliers

```python
def find_outliers_iqr(df, columns, handle_outliers=False):
    outliers = {}
    for feature in columns[:-1]:
        q1=df[feature].quantile(0.25)
        q3=df[feature].quantile(0.75)
        iqr_value=q3-q1
        lower_bound = q1 - 1.5 * iqr_value
        upper_bound = q3 + 1.5 * iqr_value
        outliers_summary = df[feature][((df[feature]<lower_bound)) | (df[feature]>upper_bound)]
        outliers[feature] = [len(outliers_summary),outliers_summary.min(),outliers_summary.max()]
        if handle_outliers:
            # Replace outliers with values based on the IQR
            df.loc[df[feature] < lower_bound, feature] = None
            df.loc[df[feature] > upper_bound, feature] = None
    return outliers
```

```python
def display_metrics(y_test, y_pred):
    mse = mean_squared_error(y_test, y_pred)
    print('Mean Squared Error  : ', mse)
    
    cm = confusion_matrix(y_test, y_pred)
    print('Confusion Matrix : \n', cm)
    total=sum(sum(cm))
    accuracy = (cm[0,0]+cm[1,1])/ total
    print ('Accuracy : ', accuracy)

    sensitivity = cm[0,0]/ (cm[0,0]+cm[0,1])
    print('Sensitivity : ', sensitivity )

    specificity = cm[1,1]/(cm[1,0]+cm[1,1])
    print('Specificity : ', specificity)    
```
